```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/cs109/content/master/imdb_top_10000.txt'
names = ['imdbID', 'title', 'year', 'score', 'votes', 'runtime', 'genres']
data = pd.read_csv(url, delimiter='\t', names=names).dropna()
print "Number of rows: %i" % data.shape[0]
```

```python
# Cleans 'mins' from runtime data
clean_runtime = [float(r.split(' ')[0]) for r in data.runtime]
data['runtime'] = clean_runtime

# Determine the unique genres
genres = set()  # set() of unique values
for m in data.genres:
    genres.update(g for g in m.split('|'))
genres = sorted(genres)

# Make a column for each genre
for genre in genres:
    data[genre] = [genre in movie.split('|') for movie in data.genres]
    
data.head()
```

```python
# Cleans year from title values
data['title'] = [t[0:-7] for t in data.title]
```

```python
# Replace zero runtime values with NaN
data.runtime[data.runtime==0] = np.nan
```

```python
# Sci-Fi DataFrame
cols = ['title', 'year', 'Sci-Fi', 'score', 'votes', 'runtime']
sci_fi = data.ix[:,cols]  # df.ix[row slice:column slice]
sci_fi.columns = ['title', 'year', 'Sci_Fi', 'score', 'votes', 'runtime']  # Rename Sci-Fi attribute to make it managable
sci_fi.head()
```

```python
# df = df.drop(df[df.score < 50].index)
true_data = sci_fi.drop(sci_fi[sci_fi.Sci_Fi == False].index)  # removes all non sci-fi films
result = true_data.sort('score', ascending=False)
result.head()

### Fact 1: The top-5 Science-Fiction films by score
```

```python
result.tail()

### Fact 2: The bottom-5 Science-Fiction films by score
```

```python
# Year groupings
scifi_year = true_data.groupby('year')
```

```python
# Series to be plotted
score_count = scifi_year.score.count()
score_mean = scifi_year.score.mean()
score_std = scifi_year.score.std()

runtime_mean = scifi_year.runtime.mean()
runtime_std = scifi_year.runtime.std()
```

```python
plt.scatter(true_data.runtime, true_data.score, lw=0, alpha=.2, color='b')
plt.xlabel("Runtime")
plt.ylabel("IMDB Rating")
# plt.xscale('log')

# Tried to visualize to see if there was any relationship between runtime and rating
```

```python
plt.scatter(true_data.votes, true_data.score, lw=0, alpha=.2, color='b')
plt.xlabel("Votes")
plt.ylabel("IMDB Rating")

# Visualized votes against rating
```

```python
plt.plot(runtime_mean.index, runtime_mean.values, 'o-',
        color='r', lw=1, label='Annual Average')
plt.fill_between(runtime_mean.index, (runtime_mean + runtime_std).values,
                 (runtime_mean - runtime_std).values, color='r', alpha=.3)
plt.scatter(true_data.year, true_data.runtime, lw=0, alpha=.2, color='k')
plt.xlabel("Year")
plt.ylabel("Runtime")
plt.title('Average Runtime for Sci-Fi Movies')
plt.legend(frameon=False)
```

```python
plt.plot(score_mean.index, score_mean.values, 'o-',
        color='r', lw=1, label='Annual Average')
plt.fill_between(score_mean.index, (score_mean + score_std).values,
                 (score_mean - score_std).values, color='r', alpha=.3)
plt.scatter(true_data.year, true_data.score, lw=0, alpha=.2, color='k')
plt.xlabel("Year")
plt.ylabel("IMDB Rating")
plt.title('Average Rating for Sci-Fi Movies')
plt.legend(frameon=False)
```
