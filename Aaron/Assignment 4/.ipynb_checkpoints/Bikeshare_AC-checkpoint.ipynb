```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
sns.set_style('white')
```

```python
data = pd.read_csv('bikeshare.csv', delimiter=',')
print "Number of rows: %i" % data.shape[0]
data.head()  # print the first 5 rows
```

```python
data.describe()
```

```python
# Group data by daily and hourly attributes
hourly = data.groupby('hr')
daily = data.groupby('dteday')
weekd = data.groupby('weekday')
workd = data.groupby('workingday')
holid = data.groupby('holiday')
# Daily series to be plotted
daily_mean_temp = daily.temp.mean()
daily_mean_hum = daily.hum.mean()
daily_mean_cnt = daily.cnt.mean()
daily_mean_casual = daily.casual.mean()
daily_mean_registered = daily.registered.mean()
### Hourly series to be plotted
hourly_mean_casual = hourly.casual.mean()
hourly_mean_registered = hourly.registered.mean()
### Weekday series to be plotted
weekd_mean_casual = weekd.casual.mean()
weekd_mean_registered = weekd.registered.mean()
```

```python
plt.scatter(daily_mean_temp, daily_mean_cnt, lw=0, alpha=.4, color='b')
plt.xlabel("Avg Daily Temperature")
plt.ylabel("Aveg Daily Total Riders")

# As expected, there is a relationship between temperature and ridership
```
