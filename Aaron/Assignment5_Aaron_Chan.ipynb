##Part 1

##Import##

```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sklearn as skl
```

##Describe##

```python
bikeshare_data = pd.read_csv('../../DAT_20_NYC/Data/bikeshare.csv')

bikeshare_data.describe()
```

## Clean

```python
bikeshare_data.drop(['instant', 'dteday'], axis=1, inplace=True)
bikeshare_data.dropna()
```

## Transform

```python
from sklearn.preprocessing import OneHotEncoder  # Import OneHotEncoder Object from Preprocessing Module

ohc = OneHotEncoder(sparse=False)  # Instantiate OHE object with sparse parameter = False
enc_weathersit = ohc.fit_transform(bikeshare_data[['weathersit']])

print enc_weathersit

print type(enc_weathersit)
```

```python
# New column labels for encoded weathersit data
weathersit_cols = ['weather1', 'weather2', 'weather3', 'weather4']

# Convert to endoded weathersit array to dataframe
encoded_dataframe = pd.DataFrame(enc_weathersit, columns=weathersit_cols)

encoded_dataframe.head()
```

```python
# Check shapes of dataframes to be merged

print bikeshare_data.shape

print encoded_dataframe.shape
```

```python
# Concatenate dataframes and drop old weathersit feature

encoded_bikeshare = pd.concat([bikeshare_data, encoded_dataframe], axis=1)

encoded_bikeshare = encoded_bikeshare.drop(['weathersit'], axis=1)

encoded_bikeshare.head()
```

## Training Model: Regression Analysis

```python
feature_cols = ['temp', 'atemp', 'hum', 'workingday', 'hr','weather1', 'weather2', 
                'weather3', 'weather4']
                
# Create X and y
X = encoded_bikeshare[feature_cols]
y = encoded_bikeshare.cnt

print type(X)
print type(y)

# Split encoded_bikeshare into training and test data
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8)

# Can also be done like so:
# n, c = bike_data.shape
# bikeshare_train = bikeshare_data.ix[:(n/2)]
# bikeshare_test = bikeshare_data.ix[(n/2)+1:]
```

```python
from sklearn.linear_model import LinearRegression

# Instantiate linear regression object from linear_model module
regr = LinearRegression()

# Fit regression against X_train and y_train
regr.fit(X_train, y_train)

# Print intercept and coefficients (beta)
print regr.intercept_
print zip(feature_cols,regr.coef_) 
```

```python
# Checking to make sure there are the same number of coefficients as (x) feature variables

len(feature_cols) == len(regr.coef_)
```

```python
# Specify R^2 value
r2_train = regr.score(X, y)

print r2_train
```

## Testing Model: Regression Analysis

```python
# Fit regression against X_test and y_test
regr.fit(X_test, y_test)

# Print intercept and coefficients (beta)
print regr.intercept_
print zip(feature_cols,regr.coef_) 
```

```python
# Specify R^2 value
r2_test = regr.score(X, y)

print r2_test
```

## Training vs. Testing

```python
print "r2_train - r2_test: %f" % float(r2_train - r2_test)
```

**The difference between the R^2 value of the training set and the testing set is -.0007, meaning r2_test had a higher correlation coefficient than r2_train. Although the difference is negligible, a reason for this might be because the testing model used a data subset that was not drawn in a randomized method, thereby potentially allowing the ordering of the data to affect the results.**

##Adding a binary rush-hour variable

```python
bikeshare_data.hr.unique()
```

```python
encoded_bikeshare['rush_hour'] = encoded_bikeshare.hr.map(lambda x: 1 if x >= 6 and x <= 9 or 
                                                          x >= 16 and x <=18 else 0)

encoded_bikeshare['rush_hour']
```

##Training Model 2: Regression Analysis

```python
feature_cols_new = ['temp', 'atemp', 'hum', 'workingday', 'hr','weather1', 'weather2', 
                'weather3', 'weather4', 'rush_hour']

# Create X and y
X = encoded_bikeshare[feature_cols_new]
y = encoded_bikeshare.cnt

# Split encoded_bikeshare into training and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8)

# Fit regression against X_train and y_train
regr.fit(X_train, y_train)

# Print intercept and coefficients (beta)
print regr.intercept_
print zip(feature_cols_new,regr.coef_) 
```

```python
r2_train_new = regr.score(X, y)

print r2_train_new
```

##Testing Model 2: Regression Analysis

```python
# Fit regression against X_test and y_test
regr.fit(X_test, y_test)

# Print intercept and coefficients (beta)
print regr.intercept_
print zip(feature_cols_new,regr.coef_) 
```

```python
r2_test_new = regr.score(X, y)

print r2_test_new
```

##Model 1 vs. Model 2

```python
print 'Model 1 Train R2: %f' % float(r2_train)
print 'Model 1 Test R2: %f' % float(r2_test)
print "Model 1 Generalization Error: %f" % float(r2_train - r2_test)

print "\n"

print 'Model 2 Train R2: %f' % float(r2_train_new)
print 'Model 2 Test R2: %f' % float(r2_test_new)
print "Model 2 Generalization Error: %f" % float(r2_train_new - r2_test_new)
```

**Including the additional feature, rush_hour, increased the correlation coefficient of Regression Model 2 and decreased the error between training and testing model outputs. Therefore, creating the rush_hour variable improved the predictive power of our model.**
