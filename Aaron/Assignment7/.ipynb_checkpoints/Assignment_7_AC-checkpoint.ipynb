```python
fn = pd.read_csv('bank.csv', sep=';')

bank = pd.DataFrame(fn)

print len(bank)

print bank.isnull().sum()
```

### 1. There are 10 variables with string values and 7 variables with numeric values. Certain string variables are binary, such asdefault, housing, and loan - exluding the 'y' target variable. Continuous variables include age, balance, duration, and compaign. Marital is a discrete, categorical variable. I'm unsure of what pdays and previous represent...

```python
# Identify binary variables and handle separately from Label Encoder

for col in bank.columns:
    print str(col) + ': ' + str(len(bank[col].unique()))
```

```python
bank_data = bank.copy()

# Drop duration feature: 'duration: last contact duration, in seconds (numeric). 

# Important note: the 'duration' attribute highly affects the output target (e.g., if duration=0 then y='no').

# Yet, the duration is not known before a call is performed. 

# Also, after the end of the call y is obviously known. 

# Thus, this input should only be included for benchmark purposes and should be discarded if the 
# intention is to have a realistic predictive model.

bank_data = bank_data.drop('duration', axis=1)

# Transform the binary variables with .map()

y_n_cols = ['default', 'housing', 'loan', 'y']

for col in y_n_cols:
    bank_data[col] = bank[col].map({'no': 0, 'yes': 1})
    
# Transform the discrte, categorical variable
    
bank_data['marital'] = bank['marital'].map({'divorced': 0, 'single': 1, 'married':  2})

bank_data.head(10)
```

```python
# Now we will take care of the other categorical variables with LabelEncoder

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

label_cols = ['job', 'education', 'contact', 'poutcome']

for col in label_cols:
    en_col = pd.DataFrame(le.fit_transform(bank[col]))
    bank_data[col] = en_col.values
    
# encode ordinal values for month feature

print bank_data['month'].unique()

months = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
month_labels = dict(zip(months, range(1,13)))

bank_data['month'] = bank['month'].map(month_labels)

bank_data.head(10)
```

### Decision-Tree Model

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.cross_validation import train_test_split

treeclf = DecisionTreeClassifier(criterion='entropy', max_features=5, random_state=10)

X = bank_data.ix[:, :-1]
y = bank_data.ix[:, -1:]

print X.shape
print y.shape

# Split data between training and testing sets

X_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)

# Fit classifier to training set

treeclf.fit(X_train, y_train)

# Explore feature importances (sorted by information gain)

feature_importance = pd.DataFrame({'feature': X.columns, 'importance': treeclf.feature_importances_}).sort('importance', ascending=False)

feature_importance.reset_index(drop=True)

feature_importance.index = feature_importance['feature']

feature_importance = feature_importance.drop('feature', axis=1)

feature_importance
```

```python
fig = plt.figure(figsize=(10,5))

plt.bar(range(0, 15), feature_importance['importance'])
plt.xticks(range(0, 15), feature_importance.index, rotation=45);
plt.title('Feature Importance');
```

```python
# Let's use the following variables for the decision tree model
tree_features = ['age', 'day', 'balance', 'month', 'job', 'y']  # removed duration feature (read UCI description)

# Define X and y

tree_data = bank_data[tree_features]

X = tree_data.loc[:, :'y']

y = tree_data[['y']]

# Split the data

from sklearn.cross_validation import train_test_split

X_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
```

```python
# Let's visualize the features we're interested in

fig = plt.figure(figsize=(15,7))

ax1 = fig.add_subplot(231)
# ax1.set_xlim(0, 1500)
ax1.hist(X['job'])
ax1.set_title('Job')

ax2 = fig.add_subplot(232)
ax2.set_xlim(0, 5000)
ax2.hist(X['balance'], bins=1000)
ax2.set_title('Balance')

ax3 = fig.add_subplot(233)
ax3.set_xlim(1, 12)
ax3.hist(X['month'], bins=12)
ax3.set_title('Month')

ax4 = fig.add_subplot(234)
ax4.set_xlim(1, 31)
ax4.hist(X['day'], bins=31)
ax4.set_title('Day')

ax5 = fig.add_subplot(235)
ax5.set_xlim(0, max(X['age']))
ax5.hist(X['age'], bins=30)
ax5.set_title('Age')

fig.tight_layout()
```

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.externals.six import StringIO
import pydot

treeclf = DecisionTreeClassifier(criterion='entropy', random_state=1, max_depth=3)

treeclf.fit(X_train, y_train)

y_pred = treeclf.predict(x_test)

print treeclf.score(x_test, y_pred)
```

```python
print treeclf.n_classes_
print treeclf.n_features_
print treeclf.n_outputs_

print treeclf.max_depth
```

### Cross-validation using bank.csv file

```python
from sklearn.cross_validation import cross_val_score

print X.shape
print y.shape

cross_val_scores = cross_val_score(treeclf, X, tree_data['y'], cv=5)

print cross_val_scores
print 'mean score: ' + str(cross_val_scores.mean())
```

```python
from sklearn import metrics

y_pred = treeclf.predict(x_test)

y_true = y_test

confusion = pd.DataFrame(metrics.confusion_matrix(y_true, y_pred))

confusion.columns = ['Pred (-)', 'Pred (+)']
                     
confusion.index = ['Actual (-)', 'Actual (+)']

confusion
```

```python
print 'Accuracy: ' + str(metrics.accuracy_score(y_true, y_pred))
print 'Precision: ' + str(metrics.precision_score(y_true, y_pred))
print 'Recall: ' + str(metrics.recall_score(y_true, y_pred))
print 'F1: ' + str(metrics.f1_score(y_true, y_pred))
```
