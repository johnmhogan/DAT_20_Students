```python
import pandas as pd
import numpy as np


import matplotlib.pyplot as plt

%matplotlib inline
#   0) Type
#  	1) *Alcohol
#  	2) *Malic acid
#  	3) *Ash
# 	4) Alcalinity of ash  
#  	5) Magnesium
# 	6) Total phenols
#  	7) Flavanoids
#  	8) Nonflavanoid phenols
#  	9) Proanthocyanins
# 	10)Color intensity
#  	11)*Hue
#  	12)OD280/OD315 of diluted wines
#  	13)*Proline

column_names = [
    'target',
    'alcohol',
    'malic_acid',
    'ash',
    'ash_alkalinity',
    'magnesium',
    'phenols',
    'flavanoids',
    'nonflavanoid_phenols',
    'proanthocyanins',
    'color_intensity',
    'hue',
    'OD280',
    'proline'
]

wine_csv = pd.read_csv('../../Data/wine.data', header=None, names = column_names)
wine_csv.head()

# wine_csv.isnull().any()
wine_csv.describe()
from sklearn import preprocessing

# wines = wine_csv[['alcohol', 'malic_acid', 'ash', 'hue', 'proline']]
wine_features = wine_csv.drop('target', axis=1)

scaled_features = preprocessing.scale(wine_features)

X = pd.DataFrame(scaled_features, columns=wine_features.columns)
y = wine_csv['target']

X.describe()
X['ash'].hist(bins=20)
X.corr()
plt.imshow(X.corr(), interpolation='nearest');
pd.scatter_matrix(wine_features, c=y, figsize=(20, 20), s=120, cmap='brg');
# Scaler applies 0 mean and unit variance:

X.describe()
# cmap_light = ListedColormap(['#FFAAAA', "#AAFFAA", "#AAAAFF"])
# cmap_bold = ListedColormap(['#FF0000', "#00FF00", "#0000FF"])
# def decision_boundary(clf, x_min, y_min, x_max, y_max, h=0.02)
#     x_range = np.arange(x_min, x_max, h)
#     y_range = np.arange(y_min, y_max, h)

#     xx, yy = np.meshgrid(x_range, y_range)

#     xx.ravel()

#     Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

#     Z = Z.reshape(xx.shape)

#     plt.figure(figsize=(18,6))
#     plt.pcolormesh(xx, yy, Z, cmap=cmap_light)
from sklearn import neighbors
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1234)

scores = []
for k in range(1, 40):
    clf = neighbors.KNeighborsClassifier(k, weights='distance')

    clf.fit(X_train, y_train);

    scores.append(clf.score(X_test, y_test))
    
plt.plot(scores)
from sklearn import metrics

y_pred = clf.predict(X_test)

conf_matrix = metrics.confusion_matrix(y_pred, y_test)

pd.DataFrame(conf_matrix, columns=['Pred1', 'Pred2', 'Pred3'], index=['Actual1', 'Actual2', 'Actual3'])
print np.array(X.columns)
from sklearn.svm import SVC
from sklearn.feature_selection import RFECV

# Create the RFE object and compute a cross-validated score.
svc = SVC(kernel="linear")

# The "accuracy" scoring is proportional to the number of correct
# classifications
rfecv = RFECV(estimator=svc, step=1, cv=StratifiedKFold(y, 2), scoring='accuracy')
rfecv.fit(X, y)

# Plot number of features VS. cross-validation scores
plt.figure()
plt.xlabel("Number of features selected")
plt.ylabel("Cross validation score (nb of correct classifications)")
plt.plot(range(1, len(rfecv.grid_scores_) + 1), rfecv.grid_scores_)
plt.show()

print rfecv.ranking_

print X.shape
print len(rfecv.ranking_)

rfecv.support_
from sklearn.feature_selection import RFE
from sklearn.cross_validation import StratifiedKFold

rfe = RFE(estimator=svc, n_features_to_select=10, step=1)

rfe.fit(X, y)
# ranking = rfe.ranking_.reshape(digits.images[0].shape)
print rfe.ranking_
```
