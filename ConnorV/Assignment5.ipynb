```python
import pandas as pd
import statsmodels.formula.api as smf
import statsmodels.api as sm
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
bike_data = pd.read_csv('/Users/cvos/DAT_20_NYC/Data/bikeshare.csv')
```

```python
bikeshare_train=bike_data[:len(bike_data)/2]
bikeshare_test=bike_data[len(bike_data)/2:]
bikeshare_train.describe()
```

```python
lin_regression_biketrain= smf.ols(formula='cnt ~ temp + atemp + hum+ workingday + hr + weathersit',data=bikeshare_train).fit()
lin_regression_biketrain.params
```

```python
lin_regression_biketrain.summary()
```

```python
lin_regression_biketest= smf.ols(formula='cnt ~ temp + atemp + hum+ workingday + hr + weathersit',data=bikeshare_test).fit()
lin_regression_biketest.params
```

```python
lin_regression_biketest.summary()
```


The r^2 value that I got for bikesharetrain was .367 
and it was .339 for bikesharetest. This lower value for the test
means that less of the bikesharetest data fits the linear model.
From looking at the values, and knowing that each dataframe spans roughly 
the same amount of time(each one year long). It would seem this difference is
largely due to variations in weather between each year. Over many years this 
would balance out, but comparing year to year gives us this issue.


```python
rush_hour_bike_share_train=bikeshare_train
rush_hour_bike_share_test=bikeshare_test
```

```python
def add_rush_hour(row):
    if (row['hr']>=6 and row['hr']<=9) or (row['hr']>=16 and row['hr']<=18):
        value=1
    else:
        value=0
    return value
```

```python
rush_hour_bike_share_train['rushhour']=bikeshare_train.apply(add_rush_hour,axis=1)
rush_hour_bike_share_test['rushhour']=bikeshare_test.apply(add_rush_hour,axis=1)
```

```python
lin_regression_rushhour_train= smf.ols(formula='cnt ~ temp + atemp + hum+ workingday + hr + weathersit+ rushhour',data=rush_hour_bike_share_train).fit()
print lin_regression_rushhour_train.params
lin_regression_rushhour_test= smf.ols(formula='cnt ~ temp + atemp + hum+ workingday + hr + weathersit+ rushhour',data=rush_hour_bike_share_test).fit()
lin_regression_rushhour_test.params
```

```python
print lin_regression_rushhour_train.summary()
lin_regression_rushhour_test.summary()
```

After adding the data for both the train and test dataframes better fit the linear model. Before the train and test regression models fit around 36.9% and 33.7%(r^2 of .369 and .337) respectively. Now they fit 50.6% and 51.5%(r^2 of .506 and .515) of the data. It is interesting to note that after adding the rush hour column the test dataset better fit the linear model. One possible explaination for this is while people used the bikeshare program for novelty in the first year, in the second year more people began using the bikes as a means of transportation to work.

## PART II

1) The r^2 value is often called the generalization error because it shows how well the model generalizes the data. A value of 1 means that it will 100% match the actual distribution of data, and 0 means the opposite. This is useful in knowing how well a model can be used to predict future data in machine learning.

2) It would seem that count is most affected by the adjusted temperature variable, which takes into account both temp and humidity.

```python
cols = ['cnt','temp','atemp','hum','workingday','hr','weathersit','rushhour']
corr_matrix = np.corrcoef(rush_hour_bike_share_train[cols].T)
sm.graphics.plot_corr(corr_matrix, xnames=cols)
```

4) There arent many strong correlations in the data set. From the matrix we can see that temp, atemp, hr and rush hour have relatively strong correlations to count. The other weak correlations are probably hurting our model as they provide much less to predict count. 

5) As we take smaller subsets of our data the p value for each variable will likely increase. When taking smaller subsections of the data it the set becomes less representative of the whole. This will increase the variation between each point.
