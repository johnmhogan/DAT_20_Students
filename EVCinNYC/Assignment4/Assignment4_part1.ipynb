```python
#GROUP PROPERTIES
#Split movies by decade - TRUNCATING THE LAST DIGIT
#So here we are truncating the trailing year, then multiplying by 10 to get decade
#print (1994//10) = 199; then mulitply by 10 to get 1990
decade = (data.year // 10) * 10

#simplifying the data - just selecting the title and year
tyd = data[['title','year']]
#creating a new field in the tyd dataframe that takes the equation and puts in the decade value
tyd['decade'] = decade
tyd.head()
```

```python
#Since decade was not in data DataFrame (was in tyd) - how does it get used here?
#mean score for all movies in each decade
decade_mean = data.groupby(decade).score.mean()
decade_mean.name = "Decade Mean"
print decade_mean
```
