```python
bikehrs = data.groupby('hr')
sumbikehrs = bikehrs.sum()
plt.plot(sumbikehrs.casual, label="casual")
plt.plot(sumbikehrs.registered, label="registered")
#registered users are peaking during rush hour and little use in the overnight hours
#casual users don't get going till about 10 am, after morning commute
```

```python
bikeweekday = data.groupby('weekday')
sumbikeweekday = bikeweekday.sum()
plt.plot(sumbikeweekday.casual)
plt.plot(sumbikeweekday.registered)
#more casual users use the bikes on the weekends; more registered users use the bikes on the week days
#registered users are more likely to use in commute
```

```python
bikemonth = data.groupby('mnth')
sumbikemonth = bikemonth.sum()
plt.plot(sumbikemonth.casual)
plt.plot(sumbikemonth.registered)
# As would expect, fewer rides in the coloder months
```

```python
bikeweather = data.groupby('weathersit')
sumbikeweather = bikeweather.sum()
plt.plot(sumbikeweather.casual)
plt.plot(sumbikeweather.registered)
#I need to relabel the indexes here - bcs shows 0 - 3 for values 1-4
# People still ride in light snow; very few horrible days
```

```python
bikeatemp = data.groupby('atemp')
sumbikeatemp = bikeatemp.sum()
plt.plot(sumbikeatemp.casual)
plt.plot(sumbikeatemp.registered)
#The warmer the air temperature, the more riders - but drops off when hit 90 degrees (45 degreeson the chart)
```

```python
bikewind = data.groupby('windspeed')
sumbikewind = bikewind.sum()
plt.plot(sumbikewind.casual)
plt.plot(sumbikewind.registered)
#Once the winds get over 10 mph the number of rides begins to decline rapidy
```

```python
bikehum = data.groupby('hum')
sumbikehum = bikehum.sum()
plt.plot(sumbikehum.casual)
plt.plot(sumbikehum.registered)
#Not sure why humidity is so spiky
#Less riding when humidity is below 25 and over 70 (bordering on rain)
```
