### Regression of Training Set

3) 
Using the `bikeshare_train` dataframe run a regression with `cnt` as our

### Add in a Rush Hour Variable

5) Create a binary variable
for rush hour defined by 6-9am & 4-6pm.

**QUESTION: Is this error an issue?**

```python
def isrush(row):
    if row['hr'] in (6,7,8,9,16,17,18):
        val = 1
    else:
        val = 0
    return val

bikeshare_train['rush'] = bikeshare_train.apply(isrush, axis=1)
bikeshare_test['rush'] = bikeshare_test.apply(isrush, axis=1)
```
