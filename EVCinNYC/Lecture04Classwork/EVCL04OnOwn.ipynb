```python
#when gets really long, good to do more like a list - 
#just style wise, easier to read and don't have to scroll 
#backk and forth - keep code nice and neat

header_row = ['mpg', 
              'cylinders', 
              'displacement',
              'horsepower', 
              'weight',
              'acceleration', 
              'model_year', 
              'origin', 
              'car_name']
```

```python
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data'
```

```python
autompg = pd.read_csv(url, header=None, na_values='?', sep='\s+')
#autompg is defining the Data Frame - this is core attribute
# header = None (like Null) because the data set has no headers so don't want to have it assign row 0 as header row 
# dataframe.columns -- can define the column headers
# because the data had weird spaces, hard to figure out; can copy into rubular (tho hard to see )
autompg.columns = header_row

#Pandas meant for datasets in a single machine; For much larger data
```
