### Our model is a linear function 

*(note the presence of the 'hat' on y)*
$$\hat{y} = \beta_1x + \beta_0$$

### In this example model is defined by two attributes: Slope and y-intercept
$$\beta_1 = \frac{\sum_{i=1}^{n}(x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^{n}(x_i - \bar{x})^2}$$

$$\beta_0 = \hat{y} - x\beta_1$$

But our model isn't an exact fit. Our original dataset deviates from the values predicted by our model by some degree of error

$$y = \hat{y} + \epsilon$$

__the error (epsilon) is often called the residual__

```python
## Here's what a raw calculation for the previous formula would look like in Python:

#  For consistency of notation I'm going to name x and y explicitly

x = diabetes_sample.blood_pressure
y = diabetes_sample.disease_progression

x_deviation_from_mean = x - x.mean()
y_deviation_from_mean = y - y.mean()

## np.dot calculates the dot product of two vectors. This is the total sum of pairwise products of each element in both arrays. 
# In other words, np.dot([a, b, c], [x, y, z]) is the same as a*x + b*y + c*z
delta_y = np.dot(x_deviation_from_mean, y_deviation_from_mean)  # The numerator of the Beta1 calc is the rise of our slope
delta_x = np.dot(x_deviation_from_mean, x_deviation_from_mean)  # The denominator of the Beta1 calc is the run of our slope

## Note: This is the same calculation. Why?
# delta_y = (x_deviation_from_mean * y_deviation_from_mean).sum()  # The numerator of the Beta1 calc.
# delta_x = (x_deviation_from_mean * x_deviation_from_mean).sum()  # The denominator of the Beta1 calc.

beta1 = delta_y / delta_x
beta0 = y.mean() - beta1*x.mean()

print beta1
print beta0
```

```python
y_hat = beta1 * x + beta0


model = pd.DataFrame({'x': x, 'y_hat': y_hat})

diabetes_sample.plot(kind='scatter', 
                     x='blood_pressure', 
                     y='disease_progression', 
                     figsize=(18, 18), 
                     fontsize=20,
                     subplots=True, 
                     marker='o', 
                     s=90);

x_range = [x.min(), x.max()]
y_hat = np.multiply(beta1, x_range) + beta0

x_df = pd.DataFrame(
{
    'Blood Pressure': [x.min(), x.max()],
    'Diab.': y_hat
})

plt.plot([x.min(), x.max()], y_hat, c='red', linewidth=2);
```
