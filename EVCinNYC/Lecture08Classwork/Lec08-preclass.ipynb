```python
#Detecing Credit Defaults - Classification

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from matplotlib.colors import ListedColormap
%matplotlib inline
url = '/Users/EVC/GA/upstream/DAT_20_Students/EVCinNYC/data/Default.csv'
df = pd.read_csv(url)
```

```python

pd.crosstab(df['student'], df['default']).rename(columns={0: 'No', 1: 'Yes'})
```

```python
#convert everythign to numeric before splitting
df.student = np.where(df.student == 'Yes', 1,0)
```

```python
#Downsample because assymmetrical 
#Need to understand strategies around that
```

```python
# downsample negative cases -- there are many more negatives than positives
indices = np.where(df.default == 0)[0]
rng = np.random.RandomState(13)
rng.shuffle(indices)
n_pos = (df.default == 1).sum()
df = df.drop(df.index[indices[n_pos:]])

df.head()

pd.crosstab(df['student'], df['default']).rename(columns={0: 'No', 1: 'Yes'})
```

```python
plt.figure(figsize=(14,8))

#scatter balance (x) and income (y)
ax1 = plt.subplot(221, title="Credit defaults by income and balance")
cm_bright = ListedColormap(['#FF0000','#00FF00'])
ax1.scatter(df.balance, df.income, c=(df.default == 1), cmap=cm_bright)
ax1.set_xlim((df.balance.min(), df.balance.max()))
ax1.set_ylim((df.income.min(), df.income.max()))
ax1.set_xlabel('balance')
ax1.set_ylabel('income')
ax1.legend(loc='upper right')

# box plots for income
ax2 = plt.subplot(222)
ax2.boxplot([df.income[df.default == 1], df.income[df.default == 0]])
ax2.set_ylim((df.income.min(), df.income.max()))
ax2.set_xticklabels(('Yes', 'No'))
ax2.set_ylabel('income')

# box plots for balance
ax3 = plt.subplot(223)
ax3.boxplot([df.balance[df.default == 0], df.balance[df.default == 1]], vert=0)
ax3.set_xlim((df.balance.min(), df.balance.max()))
ax3.set_yticklabels(('Yes', 'No'))
ax3.set_xlabel('balance')
plt.tight_layout()
```
