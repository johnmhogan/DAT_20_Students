```python

features = ['atemp', 'hum',  'workingday', 'hr', 'rushhour', 'weathersit']
target = ['cnt']

X = bikeshare[features]
y = bikeshare[target]

#ADDED IN THIS POLYOMIAL SECTION
from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree=5)
X = poly_features.fit_transform(X)

    #play with the degree and see how shape changes - more, explodes with this below
     #poly_features.fit_transform(X).shape
    #X = poly_features.fit_transform(X)

N,M = bikeshare.shape

#Can also split --  from beginning to total rows N divided by two
# HOW TO DO SPLIT NOTES
#Indexing - here first 10 rows, 
#bikeshare['hr'][:10]
#OR can use LOC function like this
#bikeshare.loc(:10, 'hr')
#sometimes defensive copy made or not depending on function
#when did polynomial above, changed to nparray so this doesn't work >> X_train = X.loc[:(N/2), :]
X_train = X[:(N/2)]
y_train = y.loc[:((N/2)-1), :]

    #FIX OLD THAT WAS X_test = X.loc[((N/2)+1):, :]
X_test = X[(N/2):]
y_test = y.loc[(N/2):, :]
```

```python
from sklearn import linear_model
linear_regressor = linear_model.LinearRegression()

linear_regressor.fit(X_train, y_train)

print "", linear_regressor.coef_
print "", linear_regressor.intercept_
```

```python
# R^2 = 1 - SSR/SST
print "Training Set R-Squared"
print linear_regressor.score(X_train, y_train)
print "Testing Set R-Squared"
print linear_regressor.score(X_test, y_test)
```

```python
#CROSS VALIDATION -
#We had used two split in half
#Could use 70% of data for training
# Or you could cut it K "folds" so use 1/K% = Testing data
# 1 - 1/K = Training Data
# do K times training samples - every time do get a series of scores
# so can look out of testing (out of sample data)
# can average all the results
# K = N ("leave one out")

# RidgeCV
# http://scikit-learn.org/stable/modules/generated/sklearn.linear_model.RidgeCV.html

from sklearn import linear_model

ridge_regressor = linear_model.RidgeCV(alphas=[0.1, 1., 10.], cv=10)
ridge_regressor.fit(X_train,y_train)

print "Training Set Ridge CV"
print ridge_regressor.score(X_train, y_train)
print "Testing Set Ridge CV"
print ridge_regressor.score(X_test, y_test)
```
