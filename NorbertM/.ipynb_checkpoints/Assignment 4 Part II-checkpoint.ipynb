```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/Users/norb/Dropbox/Data_Sets/bikeshare.csv', delimiter=',', header=0)
print "Number of rows: %i" % df.shape[0]
df.head(50)  # print the first 5 rows
```

```python
import datetime

#a = datetime.datetime.strptime('2011-01-02', '%Y-%m-%d')
#print a.weekday()

truedate = [datetime.datetime.strptime(r, '%Y-%m-%d') for r in df.dteday] #Loops through this column and extracts the runtime
df['dteday'] = truedate
```

```python
df.describe()
```

```python
df['dayofweek'] = [r.weekday() for r in df.dteday]
df.head()
```

```python
dayname = ['Sunday',
        'Monday', 
        'Tuesday', 
        'Wednesday', 
        'Thursday',  
        'Friday', 
        'Saturday'
        ]

dayname[2]
```

```python
df.head()
df['dayofweekname'] = [dayname[r] for r in df.weekday]#not sure why this is not working
```

```python
df.head(10)
```

```python
df[df.season == 1].head()
```

```python
df['atempinF']=[(r*50)*(9/5)+32 for r in df.atemp]
```

```python
df['unnormhumidity'] = [r*100 for r in df.hum]
```

```python
df.tempinF.describe()
```

```python
dfweekday = df.groupby("weekday").cnt.mean()
dfweekday.name = 'Weekday Average Count'
print dfweekday

labels = ['Sunday',
        'Monday', 
        'Tuesday', 
        'Wednesday', 
        'Thursday',  
        'Friday', 
        'Saturday'
        ]
plt.plot(dfweekday.index, dfweekday.values, 'o',
        color='b', label='Daily Average')
#plt.scatter(data.year, data.score, alpha=.04, lw=0, color='k')
plt.xticks(dfweekday.index, labels,rotation=70)
plt.xlabel("Day of the Week")
plt.ylabel("Count of Riders")
plt.ylim(ymax=220)
plt.ylim(ymin=160)
#plt.legend()
```

```python
dfhourly = df.groupby("hr").cnt.mean()
dfhourly.name = 'Hourly Average Count'
print dfhourly

plt.plot(dfhourly.index, dfhourly.values, 'o',
        color='g', label='Hourly Average')
#plt.scatter(data.year, data.score, alpha=.04, lw=0, color='k')
plt.xlabel("Time of Day")
plt.ylabel("Count of Riders")
plt.ylim(ymin=0)
#plt.legend()
```

```python
dfdaily = df.groupby("dteday").cnt.sum()
dfdaily.name = 'Daily Average Count'
print dfdaily

#plt.bar(dfhourly.index, dfhourly.values)
plt.plot(dfdaily.index, dfdaily.values)
plt.xticks(rotation=70)
plt.xlabel("Day")
plt.ylabel("Count of Riders")\
```

```python
averagedailytemp = df.groupby("dteday").atempinF.mean()
dailyridership = df.groupby("dteday").cnt.sum()
plt.scatter(averagedailytemp, dailyridership, alpha=.2, color='b')
plt.xlabel("Temp (F)")
plt.ylabel("Rider Count")
plt.ylim(ymin=0)
```

```python
averagedailyhumidity = df.groupby("dteday").unnormhumidity.mean()
dailyridership = df.groupby("dteday").cnt.sum()
plt.scatter(averagedailyhumidity, dailyridership, alpha=.2, color='g')
plt.xlabel("Humidity")
plt.ylabel("Rider Count")
plt.ylim(ymin=0)
```

```python
weathersitutation = df.groupby("weathersit").cnt.mean()
print weathersitutation

labels = ['Mostly Clear','Partly Cloudy/Misty','Little Precipitation','Heavy Precipitation']
plt.bar(weathersitutation.index, weathersitutation.values,align="center")
plt.xticks(weathersitutation.index, labels,rotation=70)
plt.ylabel("Mean Rider Count")
```

```python
season = df.groupby("season").cnt.mean()
print season

labels = ['winter','spring', 'summer','fall']
plt.bar(season.index, season.values,align="center")
plt.xticks(season.index,labels, rotation=90)
plt.ylabel("Mean Rider Count")
```
