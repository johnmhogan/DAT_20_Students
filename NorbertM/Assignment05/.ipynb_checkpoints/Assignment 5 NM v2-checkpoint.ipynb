```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

#Using Statsmodels
import statsmodels.api as sm
```

```python
df = pd.read_csv('/Users/norb/Dropbox/Data_Sets/bikeshare.csv', delimiter=',', header=0).drop('instant', axis=1)
print "Number of rows: %i" % df.shape[0]
df.tail()
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()
transformed = enc.fit_transform(df[['weathersit']].values)
#print enc.n_values_
#print enc.feature_indices_
#enc.transform([df.weathersit]).toarray()
#df.weathersit
#transformed.shape
#print transformed[2000,4]
transformed.toarray()
```

```python
columnnames = ['Clear','Cloudy','Light_Prec','Heavy_Prec']

dft = pd.DataFrame(transformed.toarray(), columns = columnnames)
dft.tail()
```

```python
#df[['weathersit']]  vs df['weathersit']
#df1 = df[['a','b']]
```

```python
df = pd.concat((df, dft), axis=1)
```

```python
df.head()
```

```python
dffeat = df[['temp','atemp','hum','workingday','hr','Clear','Cloudy','Light_Prec','Heavy_Prec']]
from sklearn import cross_validation


X_train, X_test, y_train, y_test = cross_validation.train_test_split(dffeat, df.cnt, test_size=0.5, random_state=0)
#est = sm.OLS(y, X).fit()
```

```python
X_train.shape
```

```python
# create a fitted model in one line
lm = sm.OLS(y_train,X_train).fit()

# print the coefficients
print "Coefficients:\n",lm.params,'\n'

# print the R^2
print "R^2 for Training Set is: ",lm.rsquared
```

```python
predictedval = lm.predict(X_test)
```

```python
lmtest = sm.OLS(y_test,X_test).fit()
# print the R^2
print "R^2 for Test Set is: ",lmtest.rsquared

lmprecited = sm.OLS(predictedval,X_test).fit()
print "R^2 for Predicted Set Using the Model is: ",lmprecited.rsquared
```

```python
#Create a binary variable for rush hour defined by 6-9am & 4-6pm.

def mapping(x):
    rushhour = 0
    if x>5 and x<9:
        rushhour = 1
    if x>15 and x<18:
        rushhour = 1
    return rushhour

#df['rush_hour'] = [mapping(r) for r in df.hr]
df['rush_hour'] = df.hr.apply(mapping)
```
