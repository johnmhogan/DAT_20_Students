```python
import urllib2
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd

web_scraped_data = []

def parse_table_table(soup): 
    current_row = []
    tables = soup.findAll("table")
    for table in tables:
        rows = table.find_all('tr')
        if rows:
            for row in rows:
                cols = row.find_all('td')
                column_values = [ele.text.strip() for ele in cols]
                #print column_values
                if len(column_values) == 4:
                    if "More" in column_values[2]:
                        #print "State",column_values[2].split('More ')[1].split(' data')[0]
                        current_row.append(column_values[2].split('More ')[1].split(' data')[0])#Extract state from the 3rd column
                    else:
                        current_row.append(column_values[2])
                    
                
        else:
            print "no rows found in the table - recheck table parsing logic"
    #print current_row
    web_scraped_data.append(current_row)    
            
stateid_range = np.arange(1000,57000,1000) #56 States since some are invalid entries
invalid_state_ids = [3000,7000,14000,43000,52000] #invalid

for stateid in stateid_range:
    if stateid not in invalid_state_ids:
        if stateid < 10000:
            stateid = '0' + str(stateid)
        url = r'http://quickfacts.census.gov/qfd/states/{0}.html'.format(stateid)
        #print url #used to determine which stateids are not valid
        page = urllib2.urlopen(url).read()
        soup = BeautifulSoup(page)
        soup.prettify()
        parse_table_table(soup)


#print web_scraped_data
                                    
```

```python
X = state_stats_space.as_matrix().astype(np.float) #optional but warning otherwise on later on Standard Scaler
# UserWarning: StandardScaler assumes floating point values as input, got object
```

```python
y = np.where(state_stats['DorR'] == 'D',1,0)

# This is important
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)

print "Feature space holds %d observations and %d features" % X.shape
print "Unique target labels:", np.unique(y)
```

```python
#Based on yhat blog post
from sklearn.cross_validation import KFold

def run_cv(X,y,clf_class,**kwargs):
    # Construct a kfolds object
    kf = KFold(len(y),n_folds=5,shuffle=True,random_state= 1)
    y_pred = y.copy()
    
    # Iterate through folds
    for train_index, test_index in kf:
        X_train, X_test = X[train_index], X[test_index]
        y_train = y[train_index]
        # Initialize a classifier with key word arguments
        clf = clf_class(**kwargs)
        clf.fit(X_train,y_train)
        y_pred[test_index] = clf.predict(X_test)
    return y_pred
```

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression as LogR
from sklearn.ensemble import RandomForestClassifier as RF
from sklearn.neighbors import KNeighborsClassifier as KNN

def accuracy(y_true,y_pred):
    # NumPy interprets True and False as 1. and 0.
    return np.mean(y_true == y_pred)

print "Support vector machines performance:"
y_pred = run_cv(X,y,SVC)
#print y_pred
print "%.3f" % accuracy(y, y_pred)
SVC_df = pd.DataFrame({'State':state_stats['State'].values,'y_true':y,'y_pred':y_pred})
#print SVC_df.head()
print "\nSVC incorrect predictions:"
print SVC_df[SVC_df['y_pred'] != SVC_df['y_true']]

print "\n-------------------------------------------------------------------------\n"

print "Logistic Regression performance:"
y_pred = run_cv(X,y,LogR)
#print y_pred
print "%.3f" % accuracy(y, y_pred)
LogR_df = pd.DataFrame({'State':state_stats['State'].values,'y_true':y,'y_pred':y_pred})
#print SVC_df.head()
print "\nSVC incorrect predictions:"
print LogR_df[LogR_df['y_pred'] != LogR_df['y_true']]

print "\n-------------------------------------------------------------------------\n"

print "Random forest performance:"
y_pred = run_cv(X,y,RF)
print "%.3f" % accuracy(y, y_pred)
RF_df = pd.DataFrame({'State':state_stats['State'].values,'y_true':y,'y_pred':y_pred})
#print RF_df.head()
print "\nRF incorrect predictions:"
print RF_df[RF_df['y_pred'] != RF_df['y_true']]

print "\n-------------------------------------------------------------------------\n"

print "K-nearest-neighbors performance:"
y_pred = run_cv(X,y,KNN)
print "%.3f" % accuracy(y, y_pred)
KNN_df = pd.DataFrame({'State':state_stats['State'].values,'y_true':y,'y_pred':y_pred})
#print KNN_df.head()
print "\nKNN incorrect predictions:"
print KNN_df[KNN_df['y_pred'] != KNN_df['y_true']]
```

```python
from sklearn.metrics import confusion_matrix

y = np.array(y)
class_names = np.unique(y)

confusion_matrices = [
    ( "Support Vector Machines", confusion_matrix(y,run_cv(X,y,SVC)) ),
    ( "Logistic Regressuion", confusion_matrix(y,run_cv(X,y,LogR)) ),
    ( "Random Forest", confusion_matrix(y,run_cv(X,y,RF)) ),
    ( "K-Nearest-Neighbors", confusion_matrix(y,run_cv(X,y,KNN)) ),
]

for matrix in confusion_matrices:
    print "{0} confusion matrix:\n {1}\n".format(matrix[0],matrix[1])
    
```
