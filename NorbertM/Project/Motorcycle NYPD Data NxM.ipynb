```python
data = pd.read_csv('/Users/norb/Dropbox/Data_Sets/NYPD_Motor_Vehicle_Collisions.csv')
data.shape
```

```python
from datetime import datetime
data['DATE'] = data['DATE'].apply(lambda d: datetime.strptime(d, '%m/%d/%Y'))
```

```python
data['YEAR'] = data['DATE'].apply(lambda d: d.strftime('%Y')).astype(int)
#The above could be achieved by taking the last four characters of the DATA column.
```

```python
data['VEHICLE TYPE CODE 1'].unique()
```

```python
for elem in data.columns:
    if elem.startswith('VEHICLE TYPE CODE'):
        if 'MOTORCYCLE' in data[elem].values:
            print 'Motorcycle was a vehicle involved in column', elem
```

```python
year2014 = data["YEAR"] == 2014
motorcycle_involved1 = data["VEHICLE TYPE CODE 1"]=="MOTORCYCLE"
motorcycle_involved2 = data["VEHICLE TYPE CODE 2"]=="MOTORCYCLE"
motorcycle_involved3 = data["VEHICLE TYPE CODE 3"]=="MOTORCYCLE"
motorcycle_involved4 = data["VEHICLE TYPE CODE 4"]=="MOTORCYCLE"
motorcycle_involved5 = data["VEHICLE TYPE CODE 5"]=="MOTORCYCLE"

dataofinterest = data[year2014 & (motorcycle_involved1 | motorcycle_involved2 | motorcycle_involved3 | motorcycle_involved4 | motorcycle_involved5)]
dataofinterest.count()
```

```python
#One can see that the latitude and longitude is not provided for all entries.

dataofinterest = dataofinterest[dataofinterest.LOCATION.notnull()]
dataofinterest.head()
```
