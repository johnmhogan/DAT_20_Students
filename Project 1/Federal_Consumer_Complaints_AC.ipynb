## Tranform / Clean

```python
# Since we mainly have categorical variables, let's see which features are managable in terms of the number of unique values
```

```python
cols = data.columns.tolist()

for x in cols:
    column = data[x].unique()
    print 'Unique {0}: {1}'.format(x, len(column))
```

```python
# Product, submitted via, company response, timely response appear managable - we'll drop the rest

drop_cols = ['Complaint ID', 'Sub-product', 'Issue', 'Sub-issue', 'Date received', 'Date sent to company',
             'Consumer disputed?']

complaint_data = data.drop(drop_cols, axis=1).dropna()

# Also peculiar how there are 63 states, let's investigate further...
```

```python
states = complaint_data.State.value_counts()

print states

# Top-5 states with the most complaints...

top5_states = ['CA', 'FL', 'TX', 'NY', 'GA']

#  Territories are also included in the federal dataset
```

```python
complaint_data.head()
```

```python
#  The proceeding code was created in case we wanted apply machine learning algorithms on the data
#  For the purposes of exploration and visualization, we did not have to use LabelEncoder and OneHotEncoder
```

```python
print complaint_data['Product'].unique()
print complaint_data['Company response'].unique()


unique_product = ['Debt collection' , 'Mortgage', 'Consumer loan', 'Bank account or service', 'Credit reporting',
                   'Credit card', 'Payday loan', 'Money transfers', 'Student loan', 'Other financial service', 'Prepaid card']
unique_response = ['Closed with monetary relief', 'Closed with explanation', 'Closed','In progress', 
                   'Closed with non-monetary relief', 'Untimely response', 'Closed without relief', 'Closed with relief']

print len(unique_product)
print len(unique_response)
```

```python
# First, convert categorical products to integers with Label Encoder
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

le_product = le.fit_transform(complaint_data['Product'])
le_response = le.fit_transform(complaint_data['Company response'])

product_value = range(1,12)
response_value = range(1,9)

print product_value
print response_value

product_key_mapping = dict(zip(unique_product, product_value))
response_key_mapping = dict(zip(unique_response, response_value))

complaint_data['Product'] = complaint_data['Product'].map(product_key_mapping)
complaint_data['Company response'] = complaint_data['Company response'].map(response_key_mapping)
```

```python
# Now, apply OneHotEncoder to product and company response
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse=False)

ohe_product = ohe.fit_transform(complaint_data[['Product']].dropna().values)
ohe_response = ohe.fit_transform(complaint_data[['Company response']].dropna().values)

encoded_product = pd.DataFrame(ohe_product)
encoded_response = pd.DataFrame(ohe_response)

encoded_product.columns = unique_product
encoded_response.columns = unique_response

encoded_response.head()
```

```python
#  Concatenate dataframes
```

```python
encoded_complaint = pd.concat([complaint_data, encoded_product, encoded_response], axis=1).drop(['Product', 'Company response']
                                                                                                , axis=1)
encoded_complaint.head()
```

## Explore / Visualize

```python
#  Given the top-5 states for complaints, what is the product breakdown per state?
```

```python
product_by_state = pd.crosstab(complaint_data['State'], data['Product'])

product_by_state_top5 = product_by_state.loc[top5_states]

product_by_state_top5
```

```python
product_by_state_top5.plot(kind='barh', stacked=True, figsize=(10,7), title = "Product Complaints by State")
```

```python
#  What are the outcomes/progress of each product complaint as a whole?
```

```python
response_by_product = pd.crosstab(data['Product'], data['Company response'])

response_by_product.head()
```

```python
response_by_product.plot(kind='barh', stacked=True, figsize=(10,7), title = "Company Response by Product")
```
