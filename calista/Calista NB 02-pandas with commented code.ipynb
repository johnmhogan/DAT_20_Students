```python
import pandas as pd
import numpy as np
from __future__ import division
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
pd.options.display.mpl_style = 'default'
```

```python
header_row = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'num',]
# above code should be adding headers into one row
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data'
# defines the url for below
heart_data = pd.read_csv(url, header=0, na_values='?')
# defines heart_data variable as a CSV of the url info - with the read_csv function, with no header info - since we have our own - and replaces NA values with a question mark
heart_data.columns = header_row
# delimits the column headers as the row we defined
```

```python
print heart_data.head()
#prints the top few observations of the data set
```

```python
## Lists
import random
random_list = [random.random() for i in xrange(300)]
#imports random standard library, defines random_list as a random number of items up to 300

print random_list[3:14]
print random_list[280:]
print random_list[:20]
# from the random list generated above, these commands should print items 3-13, 280-300, and 0-19

print len(random_list)
# prints the length of the random list

## DataFrame
print heart_data[3:14]
print heart_data[280:]
print heart_data[:20]
#instead of just printing the data points in a list format like above, this command prints the data points within the context of the data columns

print len(heart_data)
#length is 302 because 0-300 is 301, plus one row for the header
```
