```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cross_validation import train_test_split
from sklearn.datasets import fetch_lfw_people
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.decomposition import RandomizedPCA
from sklearn.svm import SVC
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model
from sklearn.feature_selection import f_regression
#import all the things

pd.set_option('display.width', 500)
pd.set_option('display.max_columns', 100)

%matplotlib inline
```

```python
complaints_csv = pd.read_csv('Student_loan_complaints.csv')

#testing the import
print "Number of rows: %i" % complaints_csv.shape[0]
complaints_csv.head()
```

```python
c = complaints_csv.groupby(['Issue']).size()
print c

plt.subplot(aspect=True)
plt.pie(c, labels=c.index.values, autopct='%i%%', radius = 2)
plt.title("Complaints by Issue")
```

###As you can see by the above, 69% of the student loan complaints have to do with repayment concerns.  Let's examine further breakdowns of the data to see if there are any other large majorities.
