```python
%matplotlib inline

import matplotlib.pyplot as plt
import matplotlib.dates as dts
import pandas as pd
import numpy as np

data = pd.read_csv('C:\Users\dgoodburn\DAT_20_NYC\Data\\bikeshare.csv').dropna()
print "Number of rows: %i" % data.shape[0]

data['day']=pd.to_datetime(data['dteday'])
data.head()

# import the data and convert the date to datetime64
```

```python
sumdata = data.groupby(data['day']).agg({'season' : 'mean', 'yr':'mean','holiday':'mean','weekday':'mean','workingday':'mean','temp':'mean',
                                    'hum':'mean','atemp':'mean','windspeed':'mean','casual':'sum','registered':'sum','cnt':'sum'}).reset_index()
sumdata['datindex']=sumdata.index
sumdata.head()

# group the data by day and add a column to represent the date by an integer (i realize i shouldn't need to do this but I couldn't figure out how to chart datetimes)
```

```python
plt.hist(sumdata.cnt, bins=100, color='#cccccc')
plt.xlabel("Rider distribution")

# shows number of days with each number of riders
```

```python
plt.scatter(sumdata.index, sumdata.cnt)
plt.xlabel("Day")
plt.ylabel("Riders")

# shows ridership by day over time
```

```python
plt.scatter(sumdata.index, sumdata.casual)
```

```python
plt.scatter(sumdata.index, sumdata.registered)
```

```python
import statsmodels.formula.api as smf
lm = smf.ols(formula='cnt ~ datindex', data=sumdata).fit()

X_new = pd.DataFrame({'datindex': [sumdata.datindex.min(), sumdata.datindex.max()]})
X_new.head()

preds = lm.predict(X_new)

lm.params
lm.predict(sumdata[['datindex','cnt']])
sumdata.plot(kind='scatter', x='datindex', y='cnt')

plt.plot(X_new, preds, c='red', linewidth=2)

print lm.conf_int()
print lm.pvalues
print "r2 = "+ str(lm.rsquared)

#shows relationship between ridership and day. R squared value is not high. Presumably because we do not factor in any time of seasonality.
```

```python
type1 = 'registered'
type2 = 'holiday'
lm = smf.ols(formula=type1+" ~ "+type2, data=sumdata).fit()

X_new = pd.DataFrame({type2: [sumdata.holiday.min(), sumdata.holiday.max()]})
X_new.head()

preds = lm.predict(X_new)
preds
# print the coefficients
lm.params
lm.predict(sumdata[[type2,type1]])
sumdata.plot(kind='scatter', x=type2, y=type1)

# then, plot the least squares line
plt.plot(X_new, preds, c='red', linewidth=2)

print lm.conf_int()
print lm.pvalues
print "r2 = "+ str(lm.rsquared)
```

```python
type1 = 'casual'
type2 = 'hum'
lm = smf.ols(formula=type1+" ~ "+type2, data=sumdata).fit()

X_new = pd.DataFrame({type2: [sumdata.hum.min(), sumdata.hum.max()]})
X_new.head()

preds = lm.predict(X_new)
preds
# print the coefficients
lm.params
lm.predict(sumdata[[type2,type1]])
sumdata.plot(kind='scatter', x=type2, y=type1)

# then, plot the least squares line
plt.plot(X_new, preds, c='red', linewidth=2)

print lm.conf_int()
print lm.pvalues
print "r2 = "+ str(lm.rsquared)

#doesn't seem to have any correlation which seems surprising. Again...presumably this is because there are larger contributing factors
```
