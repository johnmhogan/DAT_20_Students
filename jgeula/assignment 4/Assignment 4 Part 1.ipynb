```python
%pylab inline

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#load data, give column names, drop NAs, print # of rows and first 5 rows
names = ['imdbID', 'title', 'year', 'score', 'votes', 'runtime', 'genres']
data = pd.read_csv('https://raw.githubusercontent.com/cs109/content/master/imdb_top_10000.txt',
                   delimiter = '\t', names=names).dropna()
print "Number of rows: %i" % data.shape[0]
data.head()
```

```python
#fixing the runtime column to split number from "mins" string 
dirty = '142 mins.'
number, text = dirty.split(' ')
clean = int(number)
print number
```

```python
#packaging the above into a list comprehension and cleaning all runtime
#don't fully understand the syntax in first line, but get what it's doing
clean_runtime = [float(r.split(' ')[0]) for r in data.runtime]
data['runtime'] = clean_runtime
data.head()
```

```python
#find the unique genres
#need to research what set / update is. do understand what's going on there though
genres = set()
for m in data.genres:
    genres.update(g for g in m.split('|'))
genres = sorted(genres)
print genres
```

```python
#make a column for each genre
for genre in genres: 
    data[genre] = [genre in movie.split('|') for movie in data.genres]

data.head()
```

```python
#removes year from the title
#why 7 and not 6? two parens and 4 digit year?
#A -- because it goes up to, but not including, the last value, so it is actually 6
data['title'] = [t[0:-7] for t in data.title]
data.head()
```

```python
#splitting up movies by decade
decade =  (data.year // 10) * 10
#so using "//" will drop the float. so 195.8 will be 195

tyd = data[['title', 'year']]
tyd['decade'] = decade

tyd.head()
```

```python
# Here's info on group by: http://pandas.pydata.org/pandas-docs/dev/groupby.htmlIn [ ]:
```

```python
#mean score for all movies in each decade
#good to know you can call a function in group by (decade)
decade_mean = data.groupby(decade).score.mean()
decade_mean.name = 'Decade Mean'
print decade_mean

#this makes the line
plt.plot(decade_mean.index, decade_mean.values, 'o-',
        color='r', lw=3, label='Decade Average')
#this makes the scatterplot
plt.scatter(data.year, data.score, alpha=.04, lw=0, color='k')
#data labels
plt.xlabel("Year")
plt.ylabel("Score")
#this gives the legend for the redline
plt.legend(frameon=False)
```

```python
#compute the scatter in each year (not sure what this really means)

#just printing the head of data.groupby(decade).score doesn't make too much sense to me
grouped_scores = data.groupby(decade).score

mean = grouped_scores.mean()
std = grouped_scores.std()

#where does decade_mean come from? like i get it, but are we calling decade the variable? since when?
plt.plot(decade_mean.index, decade_mean.values, 'o-',
        color='r', lw=3, label='Decade Average')
plt.fill_between(decade_mean.index, (decade_mean + std).values,
                 (decade_mean - std).values, color='r', alpha=.2)
plt.scatter(data.year, data.score, alpha=.04, lw=0, color='k')
plt.xlabel("Year")
plt.ylabel("Score")
plt.legend(frameon=False)
```
