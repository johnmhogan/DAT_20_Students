####Here is an overview of the columns: 
 
 - instant: record index
	- dteday : date
	- season : season (1:springer, 2:summer, 3:fall, 4:winter)
	- yr : year (0: 2011, 1:2012)
	- mnth : month ( 1 to 12)
	- hr : hour (0 to 23)
	- holiday : weather day is holiday or not (extracted from http://dchr.dc.gov/page/holiday-schedule)
	- weekday : day of the week
	- workingday : if day is neither weekend nor holiday is 1, otherwise is 0.
	+ weathersit : 
		- 1: Clear, Few clouds, Partly cloudy, Partly cloudy
		- 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
		- 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
		- 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
	- temp : Normalized temperature in Celsius. The values are divided to 41 (max)
	- atemp: Normalized feeling temperature in Celsius. The values are divided to 50 (max)
	- hum: Normalized humidity. The values are divided to 100 (max)
	- windspeed: Normalized wind speed. The values are divided to 67 (max)
	- casual: count of casual users
	- registered: count of registered users
	- cnt: count of total rental bikes including both casual and registered

### Visualize Relationship Between Ridership & Other Variables

###### Looks like everyone rides more on Holidays, regardless if a casual user or a registered user.

```python
#casual users and holidays
plt.scatter(data.holiday, data.casual, lw=0, alpha=.2, color='k')
plt.xlabel("0 = holiday, 1 = regular day")
plt.ylabel("Ridership")
```

```python
#registered users and holidays
plt.scatter(data.holiday, data.registered, lw=0, alpha=.2, color='k')
plt.xlabel("0 = holiday, 1 = regular day")
plt.ylabel("Ridership")
```

###### Tempareture & Ridership -- the heat stops people from riding at a certain point like a brick wall (rather than cold where it builds up)

```python
#scatterplot for temp and ridership
#digg into .4 ish and .6ish and see whats up 

plt.scatter(data.atemp, data.cnt, lw=0, alpha=.2, color='k')
plt.xlabel("Temp")
plt.ylabel("Ridership")
```

###### Weather and ridership -- Warmer weather (but not too hot) brings out the casual riders, while registered riders will ride even in almost any weather

```python
#casual riders 

plt.figure(figsize=(8, 6), dpi=80)
plt.subplot(1, 1, 1)
plt.plot(data.casual, data.atemp, '.');
```

### Indicators for Ridership Type

```python
#hypothesis before charting: registered users use during the week, specifically ~9 and ~6
```

###### Indicators for Casual Ridership: 

Two strong indicators for casual ridership are day of the week and hour of the day. Temperature in the form of atemp (normalized feeling temperature in Celsius. the values are divided to 50 (max)) can also be an indicator. 

Casual riders use Citi bike more on the weekend, rather than weekday and ridership is highest between the hours of 12pm - 4pm. Casual riders are less likely to ride in weather that is below .4 as compared to their registered counter parts for whom the weather makes  less of a difference. 

###### Indicators for Registered Ridership: 

Two strong indicators for registered ridership are day of the week and hour of the day. 

Registered riders use Citi bike the most during the week days, with very high peaks specifically between 7 & 8am as well as 5 & 6pm (aka rush hour). This is consistent with the profile of folks who would register for Citi bike as an alternative to riding the train for their daily commute to work. 

### Improving Ridership program

Reserved bikes at 5pm, for use exclusively by registered riders. 

Looking at the hourly chart for both registered and casual, one can see that 5pm might be a bottleneck for registered riders as there are a fair amount of casual riders active at that same time. While there are certainly less casual riders during the week, this overlap could cause some pain for registered riders on their commute home.
