# Acquire and import

```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
names = ['instant','dteday','season','yr','mnth','hr','holiday','weekday','workingday','weathersit',
         'temp','atemp','hum','windspeed','casual','registered','cnt']
data = pd.read_csv('/home/oracle/GA/1_DAT_20_NYC/DAT_20_NYC/Data/bikeshare.csv', names=names, header=0).dropna()
print "Number of rows: %i" % data.shape[0]
data.head()  # print the first 5 rows
```

```python
names2 = ['dteday','season','yr','mnth','hr','holiday','weekday','workingday','weathersit',
         'temp','atemp','hum','windspeed','casual','registered','cnt']
hourly_data=data[names2]

print type(hourly_data.dteday[0])
print type(hourly_data.cnt[0])

#convert dteday to timestamp type.
hourly_data['datetime']= pd.DatetimeIndex(hourly_data.dteday)
print type(hourly_data.datetime[0])
hourly_data.describe()
```

## Clean : hourly => daily

```python
namesdaily = ['dteday','season','yr','mnth','holiday','weekday','workingday','weathersit',
         'temp','atemp','hum','windspeed','casual','registered','cnt']
cnt_daily_data=hourly_data.groupby(hourly_data.datetime).cnt.sum()
casual_daily_data=hourly_data.groupby(hourly_data.datetime).casual.sum()
registered_daily_data=hourly_data.groupby(hourly_data.datetime).registered.sum()
temp_daily_data=hourly_data.groupby(hourly_data.datetime).temp.mean()
atemp_daily_data=hourly_data.groupby(hourly_data.datetime).atemp.mean()
weekday_daily_data=hourly_data.groupby(hourly_data.datetime).weekday.mean().round()
weathersit_daily_data=hourly_data.groupby(hourly_data.datetime).weathersit.mean().round()
print type(temp_daily_data.values)
print type(temp_daily_data.index)

print '----------------------'
print 'Daily data extracts'
print '----------------------'
print cnt_daily_data.head()
print casual_daily_data.head()
print registered_daily_data.head()
print weathersit_daily_data.head()
print temp_daily_data.head()
print atemp_daily_data.head()
print weekday_daily_data.head()

print '----------------------'
print 'Daily data description'
print '----------------------'
print 'Total riders counts'
print '----------------------'
print cnt_daily_data.describe()
print '----------------------'
print 'Casual riders counts'
print '----------------------'
print casual_daily_data.describe()
print '----------------------'
print 'Registered riders counts'
print '----------------------'
print registered_daily_data.describe()
print '----------------------'
print 'Rounded weather sit'
print '----------------------'
print weathersit_daily_data.describe()
print '----------------------'
print 'Mean temp'
print '----------------------'
print temp_daily_data.describe()
print '----------------------'
print 'Mean feeling temp'
print '----------------------'
print atemp_daily_data.describe()
print '----------------------'
```

## Daily Time series graphs

The below graph shows the riderships seasonality in the year with pikes in the summer.

```python
plt.plot(cnt_daily_data.index, cnt_daily_data.values, '-',
        color='r', lw=0.3)
plt.plot(weathersit_daily_data.index, weathersit_daily_data.values*9000/3, '-',
        color='b', lw=0.1)
plt.plot(weathersit_daily_data.index, temp_daily_data.values*9000, '-',
        color='g', lw=0.3)

```

The below graph shows some correlation between the counts and the temperature with a pike at around 0.6 on the temp.

```python
plt.scatter(temp_daily_data.values, cnt_daily_data.values, lw=0, alpha=.2, color='k')
plt.xlabel("temp")
plt.ylabel("count")
```

The casual rides seem to depend on temperature with more linearity:

```python
plt.scatter(temp_daily_data.values, casual_daily_data.values, lw=0, alpha=.2, color='k')
plt.xlabel("temp")
plt.ylabel("casual count")
```

```python
plt.scatter(temp_daily_data.values, registered_daily_data.values, lw=0, alpha=.2, color='k')
plt.xlabel("temp")
plt.ylabel("registered count")
```

```python
plt.scatter(weathersit_daily_data.values, casual_daily_data.values, lw=0, alpha=.2, color='k')
plt.xlabel("weather sit")
plt.ylabel("casual count")
```

```python
plt.scatter(weathersit_daily_data.values, registered_daily_data.values, lw=0, alpha=.2, color='k')
plt.xlabel("weather sit")
plt.ylabel("registered count")
```

```python
plt.scatter(weathersit_daily_data.values, cnt_daily_data.values, lw=0, alpha=.2, color='k')
plt.xlabel("weather sit")
plt.ylabel("count")
```

Conclusion: the weather situation and temperature seem to have more influence on casual riders than on registered users.
